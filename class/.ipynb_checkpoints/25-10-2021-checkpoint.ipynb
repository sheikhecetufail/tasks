# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,21)
```

```python
y= x**2
```

```python
x
```

```python
plt.plot(x,y,'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Y=F(x)')
plt.show()
```

Creating multiplot on same canvas.... subplot

```python
plt.subplot(1,2,1) # (no.rows,no.columns) means one row and 2 columns and 1 means on ist subplot
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(x,y,'r--')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.2,2,0.4,0.8]) # gives the size of fig and not the x and y dimensions
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('subplot')
```

```python
%matplotlib inline
fig,axes=plt.subplots(2,3)
axes[0][0].plot(x,y,'r')
axes[1][2].plot(x,y)
```

```python
fig,((a1,a2),(a3,a4))=plt.subplots(2,2)
a1.plot(x,y)
a4.plot(x,y)
```

```python
fig=plt.figure(figsize=(17,6),dpi=50)
plt.plot(x,y,'r.-',label='sq')
plt.plot(x,x**3,label='cub')
plt.legend(loc=2)#loc = 1,2,3,4
```

```python
#save figures
fig.savefig('abc.png')
```

```python
plt.scatter(x,y,s=100)
```

```python
fig,axes=plt.subplots(2,3,figsize=(8,6))
axes[0][0].plot(x,y)
axes[0][0].set_ylim(0,50)
axes[0][0].grid()                    
```
