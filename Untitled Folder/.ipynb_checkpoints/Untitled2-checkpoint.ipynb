```python
import matplotlib.pyplot as plt
```

```python
import scipy
import numpy as np
```

```python
x=np.linspace(-np.pi,np.pi,100)
```

```python
# method 1
plt.plot(x,np.sin(x),label='sin')
plt.xlabel('angles in radians')
plt.ylabel('sin')
plt.title('sin graph')
plt.legend()
plt.show()
```

```python
# method2
fig,ax1= plt.subplots() #[(ax1,ax2),(ax3,ax4)] 
ax1.plot(x,np.sin(x),'-o',label="plot sin")
ax1.set_xlabel('degrees')
ax1.set_ylabel('val')
ax1.set_title('sin')
ax1.legend()
```
