```python
# Plotting lineplot
# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset Iris
flower = sns.load_dataset('iris')

# plot lineplot
sns.set_theme (style='dark' )   # this was the assignment
sns.lineplot(x = 'sepal_length',y = 'sepal_width', data = flower)#.set(title = 'Plot', xlim = 2,ylim = 1)
```

```python
flower
```

```python
sns.boxplot (x = 'species', y = 'sepal_width',data=flower)
sns.swarmplot (x = 'species', y = 'sepal_width',data=flower,color = 'red')
```

```python
a = flower[flower['species']=='setosa']['sepal_width']
```
