```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
c_d = pd.read_csv('Chilla_data2_for_plots.csv')
c_d['Research/Working experience (Float/Int) years']=c_d['Research/Working experience (Float/Int) years'].astype(float)
```

```python
#sns.boxplot(x = 'Age',y ='Research/Working experience (Float/Int) years', data = c_d)
sns.stripplot(x = 'Age',y ='Research/Working experience (Float/Int) years', data = c_d[c_d['Age'] != '46-60'])
```

```python
sns.catplot(x = 'Age',y ='Research/Working experience (Float/Int) years', data = c_d,kind = 'bar')
```
