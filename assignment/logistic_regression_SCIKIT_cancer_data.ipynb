```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```python
dataset= datasets.load_breast_cancer()
X = dataset.data
y = dataset.target
```

```python
X = pd.DataFrame(X, columns = dataset.feature_names)
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)  
```

```python
model = LogisticRegression(solver='liblinear', random_state=0)
model.fit(X_train, y_train)
```

Our model has been created. A logistic regression model has the same basic form as a linear regression model. It has an intercept and slope parameters for every feature dimension. Since we have thirty dimensions, there should be 30 slopes. Let’s have a look at these parameters.

```python
print("intercept:")
print(model.intercept_)
print("slopes:")
print(model.coef_)
```

```python
X_test
```

```python
preds = model.predict(X_test)
print(preds)
```

```python
model.score(X_test, y_test)
```

```python
plot_confusion_matrix(model, X_test, y_test)
```

## Multinomial Logistic Regression in Python

```python
dataset = datasets.load_iris()
X = dataset.data
y = dataset.target
#we skip converting X to a dataframe
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)  #in this case, you may choose to set the test_size=0. You should get the same prediction here
model = LogisticRegression(solver='liblinear',  random_state=0)
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
print(preds)
```

```python
plot_confusion_matrix(model, X_test, y_test)
```
